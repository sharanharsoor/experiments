# Visualise image

```python
def imshow(img, title):
    
    std_correction = np.asarray([0.229, 0.224, 0.225]).reshape(3, 1, 1)
    mean_correction = np.asarray([0.485, 0.456, 0.406]).reshape(3, 1, 1)
    
    # redo the normalisation done in transform
    npimg = np.multiply(img.numpy(), std_correction) + mean_correction
    
    plt.figure(figsize=(batch_size * 4, 4))
    plt.axis('off')
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.title(title)
    plt.show()
```

```python
def show_batch_images(dataloader):
    images, _ = next(iter(dataloader)) # this gives 1 batch of images
    
    outputs = model(images) # applying VGG 16 like model.
    _, pred = torch.max(outputs.data, 1)
        
    img = torchvision.utils.make_grid(images)
    imshow(img, title=[classes[x.item()] for x in pred])
    
    return images, pred
```

```python
# One thing to notice, although several things in the image, the prediction is for harmonica.
# the object that is foucsed in this image is harmonica. that means the region belongs to harmonica must be contrinuting much for classification.

images, pred = show_batch_images(evalloader)
```

# Occlusion analysis

observed that some parts of the image is contributing more to object recognition and the same helps for classification.  This will remove some pixels (Occlude) and see if the model still finds the object correctly. 

```python
# the below exp is giving, without any occlusion with 80% confidence model is predicting as harmonica.

outputs = model(images) # calling the model with images
print(outputs.shape)
outputs = nn.functional.softmax(outputs, dim=1) 
prob_no_occ, pred = torch.max(outputs.data, 1) 
prob_no_occ = prob_no_occ[0].item()
print(pred)
print(prob_no_occ)
```

```python
# occlusion function helps in removing the some pixels/cells from image and pass it to model for prediction.

def occlusion(model, image, label, occ_size=50, occ_stride=50, occ_pixel=0.5):
    # occ_pixel 0.5: this is the pixel value placed for occluded pixel.
    
    width, height = image.shape[-2], image.shape[-1]
    
    output_height = int(np.ceil((height-occ_size)/occ_stride))
    output_width = int(np.ceil((width-occ_size)/occ_stride))
    
    heatmap = torch.zeros((output_height, output_width))
    
    for h in range(0, height):
        for w in range(0, width):
            
            h_start = h*occ_stride
            w_start = w*occ_stride
            h_end = min(height, h_start + occ_size)
            w_end = min(width, w_start + occ_size)
            
            if (w_end) >= width or (h_end) >= height:
                continue
            
            input_image = image.clone().detach()
            input_image[:, :, w_start:w_end, h_start:h_end] = occ_pixel
            
            output = model(input_image)
            output = nn.functional.softmax(output, dim=1)
            prob = output.tolist()[0][label]
            
            heatmap[h, w] = prob

    return heatmap
```

```python
heatmap = occlusion(model, images, pred[0].item(), 32, 14)
```

```python
# heatmap shows the importance of each pixel. 
# heatmap representing, if a pixel is removed, what is the probablity that model would still predict the image as harmonica.
# If value is dark or near 0, that means the pixel is most important, without that probablity of the model to detect the image as harmonica is less. 
# clearly observe that all the pixels that are dark are around the harmonica in the original image. 

imgplot = sns.heatmap(heatmap, xticklabels=False, yticklabels=False,vmax=prob_no_occ)
```

# Filter visualisation

```python
alexnet = models.alexnet(pretrained=True)
print(alexnet)
```

```python
def plot_filters_single_channel_big(t):
    
    nrows = t.shape[0]*t.shape[2]
    ncols = t.shape[1]*t.shape[3]
          
    npimg = np.array(t.numpy(), np.float32)
    npimg = npimg.transpose((0, 2, 1, 3))
    npimg = npimg.ravel().reshape(nrows, ncols)
    
    npimg = npimg.T
    
    fig, ax = plt.subplots(figsize=(ncols/10, nrows/200))    
    imgplot = sns.heatmap(npimg, xticklabels=False, yticklabels=False, cmap='Greys', ax=ax, cbar=False)
```

```python
def plot_filters_single_channel(t):
    
    nplots = t.shape[0]*t.shape[1]
    ncols = 12
    nrows = 1 + nplots//ncols
    
    npimg = np.array(t.numpy(), np.float32)
    
    count = 0
    
    fig = plt.figure(figsize=(ncols, nrows))
    for i in range(t.shape[0]):
        for j in range(t.shape[1]):
            count += 1
            ax1 = fig.add_subplot(nrows, ncols, count)
            npimg = np.array(t[i, j].numpy(), np.float32)
            npimg = (npimg - np.mean(npimg)) / np.std(npimg)
            npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
            ax1.imshow(npimg)
            ax1.set_title(str(i) + ',' + str(j))
            ax1.axis('off')
            ax1.set_xticklabels([])
            ax1.set_yticklabels([])
   
    plt.tight_layout()
    plt.show()
```

```python
def plot_filters_multi_channel(t):
    
    num_kernels = t.shape[0]    
    
    num_cols = 12
    num_rows = num_kernels
    
    fig = plt.figure(figsize=(num_cols,num_rows))
    for i in range(t.shape[0]):
        ax1 = fig.add_subplot(num_rows,num_cols,i+1)
        
        npimg = np.array(t[i].numpy(), np.float32)
        npimg = (npimg - np.mean(npimg)) / np.std(npimg)
        npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
        npimg = npimg.transpose((1, 2, 0))
        ax1.imshow(npimg)
        ax1.axis('off')
        ax1.set_title(str(i))
        ax1.set_xticklabels([])
        ax1.set_yticklabels([])

    plt.tight_layout()
    plt.show()
```

```python
def plot_weights(model, layer_num, single_channel = True, collated = False):
    
    layer = model.features[layer_num]
    if isinstance(layer, nn.Conv2d):
        weight_tensor = model.features[layer_num].weight.data
        if single_channel:
            if collated:
                plot_filters_single_channel_big(weight_tensor)
            else:
                plot_filters_single_channel(weight_tensor)
        else:
            if weight_tensor.shape[1] == 3:
                plot_filters_multi_channel(weight_tensor)
            else:
                'Can only plot weights with three channels with single_channel = False'
    else:
        print('Can only visualise layers which are convolutional')
```

```python
# single_channel = False mean the image is RGB.
# passing the layer that need to visualise, the function display all the channel images. 

plot_weights(alexnet, 0, single_channel = False)
```

```python
# in this visualise each RGB channel separately. hence each channel have 3 images.

plot_weights(alexnet, 0, single_channel = True)
```

```python
# For some channels the number of channels and visualisation will be big, hence collating them and displaying all the filters in one image. 

plot_weights(alexnet, 0, single_channel = True, collated = True)
```
